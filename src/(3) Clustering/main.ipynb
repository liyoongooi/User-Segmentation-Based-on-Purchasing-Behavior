<h3> Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

from ClusteringModel import ClusteringModel
```

<h3> Loading Data

```python
df_ohe = pd.read_csv('../../data/(D) analytical_ohe.csv', index_col=0)
df_ohe_autoencoding = pd.read_csv('../../data/(E) analytical_ohe_autoencoding.csv', index_col=0)
df_minmaxscaling_ohe = pd.read_csv('../../data/(F) analytical_minmaxscaling_ohe.csv', index_col=0)
df_minmaxscaling_ohe_autoencoding = pd.read_csv('../../data/(G) analytical_minmaxscaling_ohe_autoencoding.csv', index_col=0)
df_standardscaling_ohe = pd.read_csv('../../data/(H) analytical_standardscaling_ohe.csv', index_col=0)
df_standardscaling_ohe_autoencoding = pd.read_csv('../../data/(I) analytical_standardscaling_ohe_autoencoding.csv', index_col=0)
```

<h3> Clustering

```python
models = {
    "K1": ClusteringModel("K1", df_ohe, "One-hot Encoding", "None", "k-Means"),
    "K2": ClusteringModel("K2", df_ohe_autoencoding, "One-hot Encoding + Autoencoding", "None", "k-Means"),
    "K3": ClusteringModel("K3", df_minmaxscaling_ohe, "One-hot Encoding", "Min-max Scaling", "k-Means"),
    "K4": ClusteringModel("K4", df_minmaxscaling_ohe_autoencoding, "One-hot Encoding + Autoencoding", "Min-max Scaling", "k-Means"),
    "K5": ClusteringModel("K5", df_standardscaling_ohe, "One-hot Encoding", "Standard Scaling", "k-Means"),
    "K6": ClusteringModel("K6", df_standardscaling_ohe_autoencoding, "One-hot Encoding + Autoencoding", "Standard Scaling", "k-Means"),
    
    "D1": ClusteringModel("D1", df_ohe, "One-hot Encoding", "None", "DBSCAN"),
    "D2": ClusteringModel("D2", df_ohe_autoencoding, "One-hot Encoding + Autoencoding", "None", "DBSCAN"),
    "D3": ClusteringModel("D3", df_minmaxscaling_ohe, "One-hot Encoding", "Min-max Scaling", "DBSCAN"),
    "D4": ClusteringModel("D4", df_minmaxscaling_ohe_autoencoding, "One-hot Encoding + Autoencoding", "Min-max Scaling", "DBSCAN"),
    "D5": ClusteringModel("D5", df_standardscaling_ohe, "One-hot Encoding", "Standard Scaling", "DBSCAN"),
    "D6": ClusteringModel("D6", df_standardscaling_ohe_autoencoding, "One-hot Encoding + Autoencoding", "Standard Scaling", "DBSCAN")
}

results = {
    "Model": [],
    "Encoding": [],
    "Scaling": [],
    "Clustering Algorithm": [],
    "Silhouette Score": [],
    "Calinski-Harabasz Score": []
}

for key, model in models.items():
    results["Model"].append(model.name)
    results["Encoding"].append(model.encoding)
    results["Scaling"].append(model.scaling)
    results["Clustering Algorithm"].append(model.clustering)
    results["Silhouette Score"].append(model.sil)
    results["Calinski-Harabasz Score"].append(model.ch)

df_results = pd.DataFrame(results)
df_results
```

-> Optimal Clustering Model: K1 (OHE + k-Means)

<h3> Saving Clustering Results

```python
df_results.to_csv('../../data/(J) clustering_models_performance.csv', index=False)

df_clustering = models['K1'].data
df_clustering['Cluster'] = models['K1'].labels
df_clustering.to_csv('../../data/(K) clustered_data_analytical.csv', index=True)

df_unpreprocessed = pd.read_csv('../../data/(B) cleaned_data.csv', index_col=0)
df_unpreprocessed['Cluster'] = df_clustering['Cluster']
df_unpreprocessed.to_csv('../../data/(L) clustered_data.csv', index=True)
```
